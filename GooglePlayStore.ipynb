```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('PythonPractice/googleplaystore.csv')
```

```python
data
```

```python
data.tail()
```

```python
data.Installs = data.Installs.apply(lambda x: x.strip('+'))
```

```python
data.Installs = data.Installs.apply(lambda x: x.replace(',',''))
```

```python
data.Installs = data.Installs.replace('Free', np.nan)
```

```python
data.Installs.value_counts()
```

```python
data['Installs'] = data['Installs'].apply(lambda x: x.upper())
```

```python
data['Installs']
```

```python
import seaborn as sns
```

```python
ratings = [4.1, 3.9, 4.7, 4.5, 4.3, 4.4, 3.8, 4.1, 4.4, 4.7, 4.4, 4.4, 4.2, 4.6, 4.4, 3.2, 4.7, 4.5, 4.3, 4.6, 4.0, 4.1, 4.7, None, 4.7, None, 4.8, 4.7, 4.1, 3.9, 4.1, 4.1]

# Remove None values
ratings = [rating for rating in ratings if rating is not None]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Creating the count plot using Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(x=ratings, palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Frequency of Ratings')
plt.grid(True)
plt.show()
```

```python
# Grouping the data by 'Category' and counting the number of apps in each category
category_counts = data['Category'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Apps by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Showing the pie chart
plt.show()
```

```python
category_counts = data['Category'].value_counts()

# Threshold for number of apps to show individually
threshold = 50

# Filtering out categories with fewer apps than the threshold
category_counts_filtered = category_counts[category_counts >= threshold]
others_count = category_counts[category_counts < threshold].sum()
category_counts_filtered['Others'] = others_count

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(category_counts_filtered, labels=category_counts_filtered.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Apps by Category (Top Categories)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Showing the pie chart
plt.show()
```

```python
# Grouping the data by 'Category' and counting the number of apps in each category
category_counts = data['Category'].value_counts()

# Selecting only the top 10 categories
top_categories = category_counts.head(10)

# Plotting the pie chart
plt.figure(figsize=(12, 10))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 App Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Showing the pie chart
plt.show()
```

```python
# Count the occurrences of each category
category_counts = data['Category'].value_counts()

# Slice to include only the top 10 categories
top_10_categories = category_counts[:10]

# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_10_categories, labels=top_10_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Categories of Apps')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Create a distribution plot
plt.figure(figsize=(12, 6))
sns.histplot(data['Rating'], kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Filter the data for ratings less than or equal to 7.5
filtered_data = data[data['Rating'] <= 7.5]

# Create a distribution plot
plt.figure(figsize=(10, 10))  # Set the figure size
sns.histplot(filtered_data['Rating'], kde=True, color='skyblue')
plt.title('Distribution of Ratings (Rating <= 7.5)')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
